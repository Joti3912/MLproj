```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns


from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.decomposition import PCA
```

```python
pd.set_option('display.max_columns', None)

df = pd.read_csv('train_dataset.csv')
df.describe()
```

```python
df = df.copy()
y = df[' Forest Cover Type Classes'].copy()
X = df.drop(' Forest Cover Type Classes', axis=1).copy()
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=42)
scaler = StandardScaler()
scaler.fit(X)

X_scaled = pd.DataFrame(scaler.transform(X), columns=X.columns)
X_scaled
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

PCA

```python
# Pick the number of components that explain 95% of the variance
pca = PCA(0.95)
X_pca = pca.fit_transform(X_scaled)
X_pca.shape
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, train_size=0.7, random_state=42)
model = DecisionTreeClassifier()
model.fit(X_train_pca, y_train)
model.score(X_test_pca, y_test)
```

```python

# convert one-hot columns into single columns with continuous values
df_new = df.copy()
df_new["WildernessArea"] = df_new.iloc[:, 10:14].idxmax(axis=1,numeric_only=True)
df_new["SoilType"] = df_new.iloc[:, 14:54].idxmax(axis=1,numeric_only=True)
df_new["WildernessArea"] = df_new["WildernessArea"].str.extract('\((\d+)/\d+\)', expand=False).astype(int)
df_new["SoilType"] = df_new["SoilType"].str.extract('\((\d+)/\d+\)', expand=False).astype(int)
df_new = df_new.drop(df_new.iloc[:, 10:54], axis=1)
column = df_new.pop(' Forest Cover Type Classes')
df_new[' Forest Cover Type Classes'] = column
df_new.head()
            
```

```python
y = df_new[' Forest Cover Type Classes'].copy()
X = df_new.drop(' Forest Cover Type Classes', axis=1).copy()

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=42)

# Scale the data, gives each column in X a mean of 0 and a variance of 1
scaler = StandardScaler()
scaler.fit(X_train)

X_train = pd.DataFrame(scaler.transform(X_train), columns=X_train.columns)
X_test = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns)
model = KNeighborsClassifier(n_neighbors=3, metric='euclidean')
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
y = df_new[' Forest Cover Type Classes'].copy()
X = df_new.drop(' Forest Cover Type Classes', axis=1).copy()

scaler = StandardScaler()
scaler.fit(X)

X_scaled = pd.DataFrame(scaler.transform(X), columns=X.columns)
X_scaled

pca = PCA(0.99)
X_pca = pca.fit_transform(X_scaled)
print(X_pca.shape)

X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, train_size=0.7, random_state=42)

model = DecisionTreeClassifier()
model.fit(X_train_pca, y_train)
model.score(X_test_pca, y_test)
```
